```python
#importing the libraries we will be needing
import os
import openai
import requests
from PIL import Image
from io import BytesIO

print('importing libraries complete!')
```

```python
#hiding api key
from kaggle_secrets import UserSecretsClient
user_secrets = UserSecretsClient()
secret_value_0 = user_secrets.get_secret("key")
print('complete!')
```

# Image

```python
openai.api_key = secret_value_0
response = openai.Image.create(
  prompt="painting of kings looking out a circular window to a vortex spacetime",
  n=1,
  size="256x256"
)
image_url = response['data'][0]['url']
a = requests.get(image_url)
Image.open(BytesIO(a.content))
```
